### 16) WAP to compute column-wise sum from a tuple of lists.

**Sample Input:**
```
T = ([1, 2, 3], [4, 5, 6], [7, 8, 9])
```

**Sample Output:**
```
12 15 18

```

````python
T = ([1, 2, 3], [4, 5, 6], [7, 8, 9])
sm = 0
for i in range(len(T)):
    for j in range(len(T[0])):
        sm += T[j][i]
    print(sm)    
````
