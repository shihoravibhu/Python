### 1) Character Category Counter
**Problem:** Count uppercase, lowercase, digits, and special characters in a string.

**Sample Inputs & Outputs:**
- Input: `Python3@Lab` → Uppercase: 2, Lowercase: 7, Digits: 1, Special: 1
- Input: `HELLO123` → Uppercase: 5, Lowercase: 0, Digits: 3, Special: 0
- Input: `hi@#9` → Uppercase: 0, Lowercase: 2, Digits: 1, Special: 2

```python
str = input("Enter String : ")

lowercase  , uppercase  , digits  , special = 0 , 0 , 0 , 0

for i in str:
    if(i.islower()):
        lowercase += 1
    elif(i.isupper()):
        uppercase += 1
    elif(i.isdigit()):
        digits += 1
    elif(not i.isalnum()): 
        special += 1

print(f"Uppercase: {uppercase}, Lowercase: {lowercase}, Digits: {digits}, Special: {special}")
        
```

### 7) Prefix and Suffix Removal
- `unhappy.txt` → `happy`
- `pretest.py` → `test`
- `unwanted.doc` → `wanted`

```python
s = input("Enter String : ")

prefixes = ["pre", "un"]
suffixes = [".txt", ".py", ".doc"]

# remove prefix
for p in prefixes:
    if s.startswith(p):
        s = s.removeprefix(p)
        break

# remove suffix
for suf in suffixes:
    if s.endswith(suf):
        s = s.removesuffix(suf)
        break

print(s)
```

### 10) Toggle Case (without using str.swapcase())
- `PyThOn` → `pYtHoN`
- `HELlo` → `helLO`
- `Python` → `pYTHON`

```python
str = input("Enter String : ")

# print(str.swapcase())

s = ""

for i in str:
    if(i.isupper()):
        s += i.lower()
    elif(i.islower()):
        s += i.upper()
    else:
        s += i

print(s)
```

### 11) Username Validation
#### valid username rules:
1. only alphabets and digits
2. does not start with a digit
3. length ≥ 6

- `User123` → Valid
- `123User` → Invalid
- `Us@12` → Invalid

```python
username = input("Enter UserName : ")

if(username.isalnum() and not username[0].isdigit() and len(username) >= 6):
    print(f"{username} Is Valid")
else:    
    print(f"{username} Is Invalid")    
```
