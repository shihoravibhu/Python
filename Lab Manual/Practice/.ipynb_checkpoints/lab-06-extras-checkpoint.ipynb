### `Problem 1:` Combine two lists index-wise(columns wise)

Write a program to add two lists index-wise. Create a new list that contains the 0th index item from both the list, then the 1st index item, and so on till the last element. any leftover items will get added at the end of the new list.

`Given List:`
```
list1 = ["M", "na", "i", "¥@"]
list2 = ["y", "me", "s", "$#"]
```

`Output:`
```
[['M','y'], ['na', me'], ['i', 's'], ['¥@', '$#']]
```

````python
list1 = ["M", "na", "i", "¥@"]
list2 = ["y", "me", "s", "$#"]

list3 = []

for i in range(len(list2)):
    list4 = []
    list4.append(list1[i])
    list4.append(list2[i])
    list3.append(list4)

print(list3)
````

### `Problem 2:` Add new item to list after a specified item
Write a program to add item 7000 after 6000 in the following Python List
```
list1 = [10, 20, [300, 400, [5000, 6000], 500], 30, 40]
```
`Output:`
```
[10, 20, [300, 400, [5000, 6000, 7000], 500], 30, 40]
```

````python
list1 = [10, 20, [300, 400, [5000, 6000], 500], 30, 40]

# list1[2][2].append(7000)
# print(list1)

# or

def append_after_value(lst, target, new_val):
    for item in lst:
        if isinstance(item, list):  # or type(item) = list
            append_after_value(item, target, new_val)
        else:
            if item == target:
                lst.append(new_val)

append_after_value(list1, 6000, 7000)

print(list1)
````

### `Problem 4:` Running(Cummulative) Sum on list
Write a program to print a list after performing running sum on it.

i.e:

`Input:`
```
list1 = [1,2,3,4,5,6]
```
`Output:`
```
[1,3,6,10,15,21]
```

````python
list1 = [1, 2, 3, 4, 5, 6]

result = []

result.append(list1[0])

for i in range(1,len(list1)):
    result.append(result[i-1]+list1[i])

print(result)
````

### `Problem 5:` You are given a list of integers. You are asked to make a list by running through elements of the list by adding all elements greater and itself.

i.e. Say given list is `[2,4,6,10,1]`
resultant list will be `[22,20,10,23]`.

For 1st element `2` ->> these are greater `(4+6+10)` values and `2` itself so on adding becomes `22`.

For 2nd element `4` ->> greater elements are `(6, 10)` and `4` itself, so on adding `20`

like wise for all other elememts.

`[2,4,6,10,1]`-->`[22,20,16,10,23]`




````python
L = [2, 4, 6, 10, 1]
result = []

result = L.copy()

for i in range(len(L)):
    
    for j in L:
        
        if L[i] < j:
            result[i] += j
    

print(result)
````

### `Problem 6:` Find list of common unique items from two list. and show in increasing order

`Input`

```
num1 = [23,45,67,78,89,34]
num2 = [34,89,55,56,39,67]
```

`Output:`
```
[34, 67, 89]
```


````python
num1 = [23, 45, 67, 78, 89, 34, 67]
num2 = [34, 89, 55, 56, 39, 67, 67]

common = []

for i in num1:
    if(i in num2 and common.count(i) == 0):
        common.append(i)
common.sort()
print(common)
````

### `Problem 7:` Sort a list of alphanumeric strings based on product value of numeric character in it. If in any string there is no numeric character take it's product value as 1.

`Input:`
```
['1ac21', '23fg', '456', '098d','1','kls']
```

`Output:`
```
['456', '23fg', '1ac21', '1', 'kls', '098d']

```

````python
L = ["1ac21", "23fg", "456", "098d", "1", "kls"]

def product_digit(s):
    product = 1
    for ch in s:
        if ch.isdigit():
            product *= int(ch)
    return product

L.sort(key = product_digit,reverse=True) # its sort method doesnt return

print(L)
````

### `Problem 9:` Convert Character Matrix to single String using string comprehension.

**Example 1:**

Input:
```bash
[['c', 'a', 'm', 'p', 'u', 'x'], ['i', 's'], ['b', 'e', 's', 't'], ['c', 'h', 'a', 'n', 'n', 'e', 'l']]
```

Output:
```bash
campux is best channel
```

````python
L = [
    ["c", "a", "m", "p", "u", "x"],
    ["i", "s"],
    ["b", "e", "s", "t"],
    ["c", "h", "a", "n", "n", "e", "l"],
]
s = ""

for i in L:
    if type(i) == list:
        s += "".join(i)
    s += " "
    
print(s)
````

### `Problem 10:` Add Space between Potential Words.

**Example:**

Input:

```bash
['campusxIs', 'bestFor', 'dataScientist']
```

Output:
```bash
['campusx Is', 'best For', 'data Scientist']
```

````python
# initializing list
test_list = ["campusxIs", "bestFor", "dataScientist"]

res = []

for s in test_list:
    for i in range(len(s)):
        if s[i].isupper():
            word = s[i:]
            s.removesuffix(word)
            s += " " + word
            res.append(s)

print(res)
````

### `Problem 11:` Write a program that can perform union operation on 2 lists

**Example:**

Input:

```bash
[1,2,3,4,5,1]
[2,3,5,7,8]
```

Output:
```bash
[1,2,3,4,5,7,8]
```

````python
L1 = [1, 2, 3, 4, 5, 1]
L2 = [2, 3, 5, 7, 8]

union = []

for i in L1:
    if i not in union:
        union.append(i)
        
for i in L2:
    if i not in union:
        union.append(i)

print(union)
````

### `Problem 12:` Write a program that can find the max number of each row of a matrix

**Example:**

Input:

```bash
[[1,2,3],[4,5,6],[7,8,9]]
```

Output:
```bash
[3,6,9]
```

````python
L = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
result = []

for li in L:
    result.append(max(li))

print(result)
````

### `Problem 14:` Write a list comprehension that can transpose a given matrix

matrix = [<br>
  [1,2,3],<br>
          [4,5,6],<br>
          [7,8,9]<br>]<br>

[1, 4, 7]<br>
[2, 5, 8]<br>
[3, 6, 9]<br>

````python
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

result = [[matrix[i][j] for i in range(len(matrix))] for j in range(len(matrix[0]))]

print(result)
````

### `Problem 15:` Write a list comprehension that can flatten a nested list

Input<br>
matrix = [<br>
  [1,2,3],<br>
          [4,5,6],<br>
          [7,8,9]<br>]<br>

Output:<br>
[1, 2, 3, 4, 5, 6, 7, 8, 9]

````python
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

flatten = [matrix[i][j] for i in range(3) for j in range(3)]

print(flatten)
````
