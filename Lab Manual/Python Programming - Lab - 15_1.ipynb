### 03) WAP to create Circle class with area and perimeter function to find area and perimeter of circle.

```python
import math

class Circle :
    def __init__(self,r) : 
        self.r = r

    def area(self):
        return math.pi * self.r * self.r
    def perimeter(self):
        return 2 * math.pi * self.r


c1 = Circle(7)
c1.area()
c1.perimeter()
```

### 05) Create a bank account class with methods to deposit, withdraw, and check balance.

```python
class Bank:
    def __init__(self):
        self.bal = 5000
        
    def deposit(self,amount):
        self.bal += amount
        print("Deposit : ",amount)
        
    def withdraw(self,amount):
        self.bal -= amount
        print("Withraw : ",amount)
        
    def displayBal(self):
        print("Current : ",self.bal)

b1 = Bank()
b1.deposit(100)
b1.withdraw(100)
b1.displayBal()
    
```

### 06) Create a class for managing inventory that includes attributes such as item name, price, quantity, and methods to add, remove, and update items.

```python
class inventory:
    def __init__(self,name,price,quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def add(self,quantity):
        self.quantity +=  quantity

    def remove(self,quantity):        
        self.quantity -=  quantity

    def update(self,name,price,quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

i1 = inventory("vi",9999,10)
i1.add(10)
print(i1.quantity)
i1.remove(5)
print(i1.quantity)
```

### 09) Define Time class with hour and minute as data member. Also define addition method to add two time objects.

```python
class Time:
    def __init__(self,h,m,s):
        self.h = h
        self.m = m
        self.s = s

    def add(self,t2):
        total_h = self.h + t2.h
        total_m = self.m + t2.m
        total_s = self.s + t2.s

        if(total_s >= 60):
            total_m += total_s // 60
            total_s = total_s % 60

        if(total_m >= 60):
            total_h += total_m // 60
            total_m = total_m % 60

        total_h = total_h % 12

        return Time(total_h,total_m,total_s)

t1 = Time(1,20,50)
t2 = Time(1,50,50)
t3 = t1.add(t2)
print(f"{t3.h} : {t3.m} : {t3.s}")
```
