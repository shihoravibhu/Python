### 13) WAP to print sum of digits of given number.

```python
n = int(input("Enter Number : "))

sum = 0

while(n != 0):   
    sum += n%10
    n //= 10

print(sum)
```

### 14) WAP to check whether the given number is Palindrome or not.

```python
n = int(input("Enter Number : "))
original_num = n
# n2 = n[::-1]

# if(n == n2):
#     print(n,"Is Palindrome")
# else:
#     print(n,"Is Not Palindrome")
#     print(n,"Is Not Palindrome")

#or

rev = 0

while n != 0:
    rev = rev * 10 + (n % 10)
    # rev*=10+ (n % 10)=> rev=10( rev+(n % 10))
    n //= 10

if(rev == original_num):
    print(original_num,"Is Palindrome")
else:
    print(original_num,"Is Not Palindrome")
```

### 15) WAP to check whether the given number is an Armstrong Number or not.

```python
n = input("Enter Number : ")

power = len(n)

n = int(n)

original_num = n

sum = 0

while(n != 0):
    sum += (n%10) ** power
    n //= 10

if(sum == original_num):
    print(original_num,"Is Armstrong")
else:
    print(original_num,"Is Not Armstrong")
```

### 16) WAP to print all the perfect numbers between 1 to n.

```python
n = int(input("Enter Number : "))

for i in range(2,n):
    sum = 0
    for j in range(1,i):
        if(i % j == 0):
            sum += j
    else:
        if(sum == i):
            print("Perfect Num : ",i)
```
