### 03) WAP to accept three integers and display the numbers, their sum, and average.
#### Input: 10 20 30

```python
num1 = int(input("Enter Number 1 : "))
num2 = int(input("Enter Number 2: "))
num3 = int(input("Enter Number 3 : "))


print(num1 , num2 , num3)

print("Sum : " , (num1 + num2 + num3))

print("Avg : ",(num1 + num2 + num3)/3)
```

### 09) WAP to display following table of items with proper alignment.
#### Output :
| Sr No | Name        | Subject | Grade | Percentage |
| ----: | ----------- | ------- | :---: | ---------: |
|     1 | Nisha Patel | Math    |   A   |         92 |
|     2 | Aarav Modi  | Science |   B+  |         85 |
|     3 | Jiya Shah   | English |   A+  |         96 |

```python
print(f"{'SrNo':^15}{'Name':15}{'Subject':15}{'Grade':^15}{'Percentage':>15}")
print(f"{1:^15}{'Nisha Patel':15}{'Math':15}{'A':^15}{92:>15}")
print(f"{2:^15}{'Aarav Modi':15}{'Science':15}{'B+':^15}{85:>15}")
print(f"{3:^15}{'Jiya Shah':15}{'English':15}{'A+':^15}{96:>15}")
```

### 10) WAP to accept a float number and display with 2 decimals, 3 decimals, and width 10.
#### Input : 37.2567
#### Output :
##### 2 decimals: 37.26
##### 3 decimals: 37.257
##### Width 10:      37.26

```python
num = float(input("Enter Num : "))

print("2 decimal" , f"{num : .2f}")
print("3 decimal" , f"{num : .3f}")
print("Width 10" , f"{num : 10.2f}")
```

### 12) WAP to accept date in dd mm yyyy format and display in multiple formats.
#### Input : 01 12 2025
#### Output :
##### 01/12/2025
##### 2025-12-01

```python
day , month , year = input("Enter Date in dd mm yyyy formate.").split()

print(day + "/" + month + "/" + year)

print(year + "-" + month + "-" + day)
```

### 15) WAP to get the distance from user into kilometer, and convert it into meter, feet, inches and centimeter.

```python
km = int(input("Enter distance (km): "))

m = km * 1000

cm = m * 100

inches = cm / 2.54  # 1 inch = 2.54 cm

ft = inches / 12 # 1 foot = 12 inches

print("Meters     :", m)
print("Centimeters:", cm)
print("Inches     :", inches)
print("Feet       :", ft)
```
