### 01) Write a function to calculate BMI given mass and height. (BMI = mass/h**2)

```python
mass , h = input("Enter Mass And h : ").split(" ")
mass = float(mass)
h = float(h)

def calBMI(mass , h):
    BMI = mass / h ** 2

    return BMI

res = calBMI(mass,h)

print(res)
```

### 02) Write a function that add first n numbers.

```python
n = int(input("Enter Number n : "))

def printFirstN(n):
    add = 0
    for i in range(1,n+1):
        add += i
        
    return add

print(printFirstN(n))
```

### 03) Write a function that returns 1 if the given number is Prime or 0 otherwise.

```python
n = int(input("Enter Number : "))

def isPrime(n):
    for i in range(2,int(n**0.5)+1):
        if(n % i == 0):
            return 0

    return 1

print(isPrime(n))     
```

### 04) Write a function that returns the list of Prime numbers between given two numbers.

```python
a , b = input("Enter Range a And b : ").split(" ")
a = int(a)
b = int(b)

def isPrime(n):
    for i in range(2,int(n**0.5)+1):
        if(n % i == 0):
            return 0

    return 1

def primeRange(a,b):
    li = []

    for i in range(a,b+1):
        if(isPrime(i) == 1):
            li.append(i)

    return li

print(primeRange(a,b))
```

### 05) Write a function that returns True if the given string is Palindrome or False otherwise.

```python
s = input("Enter String : ")

checkPalindrome = lambda x : x == x[::-1]

print(checkPalindrome(s))
```

### 07) Write a function to calculate the sum of the first element of each tuples inside the list.

```python
li = [(1,2),(3,4),(5,6)]

def sumListTuple(li):
    s = 0
    for i in li:
        s += i[0]
    return s

print(sumListTuple(li))
```

### 08) Write a recursive function to find nth term of Fibonacci Series.

```python
n = int(input("Enter nTh Term : "))

# 0 1 1 2 3 5 8 13

def fibonacciNth(n):
    if(n == 0 or n == 1):
        return n
    return fibonacciNth(n-2) + fibonacciNth(n-1)
        
fibonacciNth(n-1)
```

### 10) Write a function to get the sum of the scores ending with zero.
#### Example : scores = [200, 456, 300, 100, 234, 678] 
#### Ans = 200 + 300 + 100 = 600

```python
scores = [200, 456, 300, 100, 234, 678]

def endWithZero(scores):
    s = 0
    for i in scores:
        if(i % 10 == 0):
            s += i
    return s

print(endWithZero(scores))       
```

### 11) Write a function to invert a given Dictionary.
#### hint: keys to values & values to keys
#### Before : {'a': 10, 'b':20, 'c':30, 'd':40}
#### After : {10:'a', 20:'b', 30:'c', 40:'d'}

```python
di = {'a': 10, 'b':20, 'c':30, 'd':40}

def invert(di):
    return { v : k for k , v in di.items()}

print(invert(di))
```

### 12) Write a function that returns the number of uppercase and lowercase letters in the given string. 
#### example : Input : s1 = AbcDEfgh ,Ouptput : no_upper = 3, no_lower = 5

```python
s1 = "AbcDEfgh"

def countUpperCase(s1):
    no_upper = 0
    no_lower = 0
    for ch in s1:
        if(ch.isupper()):
            no_upper += 1
        else:
            no_lower += 1
            
    return no_upper , no_lower

print(countUpperCase(s1))        
```

### 13) Write a lambda function to get smallest number from the given two numbers.

### 14) For the given list of names of students, extract the names having more that 7 characters. Use filter().

```python
names = ["hey" , "python" , "ok" , "okokokk" , "parisssss"]

res = list(filter(lambda x : len(x) >= 7,names))

res
```

### 15) For the given list of names of students, convert the first letter of all the names into uppercase. use map().

```python
names = ["hey" , "python" , "ok" , "okokokk" , "parisssss"]

res = list(map(lambda x : x.title(),names))

res
```
